## 2. Comparación de categorías

```python
#Leemos los datos 
df2 = pd.read_csv('./datasets_top_4.csv')
df2
```

```python
# Dividir el DataFrame en dos partes
n = len(df2) // 2
df2_parte1 = df2.iloc[:n]
df2_parte2 = df2.iloc[n:]
```

```python
import matplotlib.patches as mpatches
```

```python
# Crear una figura y dos subgráficos con tamaño ajustado
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Crear el primer gráfico
df2_parte1.plot.barh(
    stacked=True,                      # Apilar las barras
    color=['#000086', '#5396F4', '#A1DBE8', '#DADADA'],  # Colores de las categorías
    xlabel='Porcentaje',
    ylabel='Dataset',
    ax=ax1  # Dibujar en el primer subgráfico
)

# Configurar los nombres de los datasets en el eje y para la parte 1
ax1.set_yticklabels(df2_parte1['dataset'])

# Crear el segundo gráfico de barras apiladas horizontalmente en el segundo subgráfico
df2_parte2.plot.barh(
    stacked=True,                      # Apilar las barras
    color=['#000086', '#5396F4', '#A1DBE8', '#DADADA'],  # Colores de las categorías
    xlabel='Porcentaje',
    ax=ax2  # Dibujar en el segundo subgráfico
)

# Configurar los nombres de los datasets en el eje y para la parte 2
ax2.set_yticklabels(df2_parte2['dataset'])

# Eliminar la leyenda de colores de cada subgráfico individual
ax1.get_legend().remove()
ax2.get_legend().remove()

# Crear una leyenda personalizada para los colores de las categorías en la figura
legend_labels = ['Categoría1', 'Categoría2', 'Categoría3', 'Otros']
legend_colors = ['#000086', '#5396F4', '#A1DBE8', '#DADADA']
legend_patches = [mpatches.Patch(color=color, label=label) for color, label in zip(legend_colors, legend_labels)]

# Agregar la leyenda a la figura como un texto adicional
fig.legend(handles=legend_patches, loc='lower center', bbox_to_anchor=(0.5, -0.05), ncol=4)

# Configurar el título de la figura y hacerlo más grande
fig.suptitle('Como se distribuyen las categorías dentro de los datasets?', fontsize=16)

# Posicionar los nombres de los datasets a la derecha en el gráfico de la derecha
ax2.yaxis.tick_right()

# Mostrar la figura con ambos subgráficos y la leyenda
plt.show()
```

## 3. Surfistas

```python
# Cargar los datos
dataSurf = pd.read_csv('viajes_surfistas.csv')
dataSurf
```

```python
# Contar el número de surfistas por país de procedencia
surfistas_por_pais = dataSurf['homecountry'].value_counts()
surfistas_por_pais
```

```python
# Calcular el porcentaje de surfistas por país
porcentaje_surfistas_por_pais = (surfistas_por_pais / surfistas_por_pais.sum()) * 100
porcentaje_surfistas_por_pais
```

```python
# Seleccionar los cuatro países con los porcentajes más altos
top_paises = porcentaje_surfistas_por_pais.nlargest(4)
top_paises
```

```python
# Calcular la suma de los porcentajes de los demás países
otros_paises = porcentaje_surfistas_por_pais.drop(top_paises.index).sum()
otros_paises
```

```python
# Resetear el índice del DataFrame
otros_serie = pd.Series({ 'Otros': otros_paises })
otros_serie
```

```python
# Eliminar la columna 'count'
catFinales = pd.concat([top_paises, otros_serie], axis=0)
catFinales
```

```python
import matplotlib.pyplot as plt

# Lista de colores personalizados
colores_personalizados = ['#FBB4AE', '#CCEBC5', '#FED9A6', '#E5D8BD', '#F2F2F2']

# Crear un gráfico de tarta con colores personalizados
catFinales.plot.pie(autopct='%1.1f%%',
                    figsize=(10, 6),
                    legend=True,
                    startangle=90,
                    colors=colores_personalizados,
                    fontsize='13',
                   )

# Ajustar el tamaño de la letra del título
plt.title('Porcentaje de Surfistas por País de Procedencia', fontsize='17')

# Ajustar la posición de la leyenda y cambiar su título y tamaño de fuente
plt.legend(title="Pais de Surfistas", loc="upper right", title_fontsize='13', bbox_to_anchor=(1.5, 1))

# Mostrar el gráfico
plt.show()
```

## 4. Ventas

```python
dataVentas = pd.read_csv('tienda_ventas.csv')
dataVentas
```

```python
# Obtener las 5 tiendas (family) con la suma más alta de ventas
top_families = dataVentas.groupby('family')['sales'].sum().nlargest(5).index
top_families
```

```python
# Convertir la columna 'date' a tipo datetime para agrupar por mes
dataVentas['date'] = pd.to_datetime(dataVentas['date'])

# Agregar una columna para el mes y el año
dataVentas['year_month'] = dataVentas['date'].dt.to_period('M')

# Filtrar las ventas solo de las 5 tiendas principales
dataVentas_top_families = dataVentas[dataVentas['family'].isin(top_families)]
dataVentas_top_families
```

```python
# Agrupar y sumar las ventas por tienda en intervalos de 1 mes
ventas_mensuales = dataVentas_top_families.groupby(['family', 'year_month'])['sales'].sum().reset_index()

# Convertir 'year_month' a una fecha de tipo datetime para la gráfica
ventas_mensuales['year_month'] = ventas_mensuales['year_month'].dt.to_timestamp()
ventas_mensuales
```

```python
# Crear la gráfica de líneas
plt.figure(figsize=(14, 7))

for family in top_families:
    family_data = ventas_mensuales[ventas_mensuales['family'] == family]
    plt.plot(family_data['year_month'], family_data['sales'], label=family)

plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.title('Ventas mensuales por tienda')
plt.legend(title='Family')
plt.grid(True)
plt.show()
```
